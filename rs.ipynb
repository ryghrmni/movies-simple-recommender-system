```python
import pandas as pd
import numpy as np
from math import e
from tkinter import *
import tkinter as tk
from difflib import SequenceMatcher
```

```python
def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()
```

```python
movies = pd.read_csv("movies.csv")
ratings = pd.read_csv("ratings_sample.csv")
```

```python
movies.head()
```

```python
movies["year"] = movies.title.str.extract('(\(\d\d\d\d\))', expand=False)
movies["year"] = movies.year.str.extract("(\d\d\d\d)", expand=False)
movies["title"] = movies.title.str.replace('(\(\d\d\d\d\))', '', regex=True)
movies["title"] = movies["title"].apply(lambda x: x.strip())
```

```python
movies.head()
```

```python
movies["genres"] = movies.genres.str.split(r"|")
```

```python
movies.head()
```

```python
movies_with_genres = movies.copy()
glist = []
for l in movies.genres:
    for g in l:
        if not g in glist:
            glist.append(g)
            movies_with_genres[g] = 0.0
```

```python
movies_with_genres.head()
```

```python
p = 0
for l in movies.genres:
    for g in l:
        movies_with_genres.at[p, g] = 1.0
    p += 1
```

```python
movies_with_genres.head()
```

```python
ratings = ratings.drop("timestamp", axis='columns')
```

```python
ratings.head()
```

```python
def add(title, rating):
    ifTitle = movies["title"].str.lower().isin([title.lower()])
    if movies[ifTitle].empty:
        for w in movies["title"].str.lower():
            if similar(title.lower(), w) >= 0.8:
                add_label.config(text= "Added!")
                userInput.append(
                    {'title': w, 'rating': int(rating)}
                    )
                break
            else:
                add_label.config(text= "Movie Not Found!")
    else:
        userInput.append(
            {'title': title, 'rating': int(rating)}
        )
        add_label.config(text= "Added!")
```

```python
def dfInputMovies():
    global inputMovies
    inputMovies = pd.DataFrame(userInput)
    inputMovies
```

```python
def dfInputId():
    global inputMovies, inputid
    inputid = movies[movies['title'].isin(inputMovies['title'].tolist())]
    inputMovies = pd.merge(inputid, inputMovies)
    inputMovies = inputMovies.drop('genres', axis='columns').drop('year', axis='columns')
    inputMovies
```

```python
def dfUserMovies():
    global userMovies
    userMovies = movies_with_genres[movies_with_genres['movieId'].isin(inputMovies['movieId'].tolist())]
    userMovies
```

```python
def dfGenresTable():
    global userMovies, userGenresTable
    userMovies = userMovies.reset_index(drop = True)
    userGenresTable = userMovies.drop("movieId", axis='columns').drop("title",axis='columns').drop("genres", axis='columns').drop("year", axis='columns')
    userGenresTable
```

```python
def dfUserProfile():
    global userProfile
    userProfile = userGenresTable.transpose().dot(inputMovies["rating"])
    userProfile
```

```python
def dfGenre_Table():
    global genre_table
    genre_table = movies_with_genres.set_index(movies_with_genres["movieId"])
    genre_table = genre_table.drop("movieId", axis='columns').drop("title", axis='columns').drop("genres", axis='columns').drop("year", axis='columns')
    genre_table.head()
```

```python
def dfRecommendation():
    global recommendation
    recommendation = ((genre_table*userProfile).sum(axis = 1))/(userProfile.sum())
    recommendation.head()
```

```python
def dfReSort():
    global recommendation
    recommendation = recommendation.sort_values(ascending= False)
    recommendation.head()
```

```python
def result():
    dfInputMovies()
    dfInputId()
    dfUserMovies()
    dfGenresTable()
    dfUserProfile()
    dfGenre_Table()
    dfRecommendation()
    dfReSort()
    
    result_text.insert(tk.END, movies.loc[movies["movieId"].isin(recommendation.head(20).keys())])
```

```python
def filled():
    if title_entry.get() and rating_entry.get():
        add_button.config(state= NORMAL)
    else:
        add_button.config(state= DISABLED)
    root.after(1, filled)
```

```python
def user_added():
    if len(userInput) > 0:
        confirm_button.config(state= NORMAL) 
    root.after(1, user_added)
```

```python
root = tk.Tk()
root.title("Movies Recommendation")
root.geometry("1000x500")

userInput = []

title_label = Label(root, text="Movie Title:")
title_entry = Entry(root)
rating_label = Label(root, text="Your Rating (from 0 to 5):")
rating_entry = Entry(root)
add_button = Button(root, text="Add", command= lambda: add(title_entry.get(), rating_entry.get(),), state= DISABLED)
add_label = Label(root, text="")
confirm_button = Button(root, text="Confirm", command= result, state= DISABLED)
result_text = Text(root, width=100, height=20)

title_label.pack()
title_entry.pack()
rating_label.pack()
rating_entry.pack()
add_button.pack()
add_label.pack()
confirm_button.pack()
result_text.pack()

root.after(1, user_added)
root.after(1, filled)
root.mainloop()
```
